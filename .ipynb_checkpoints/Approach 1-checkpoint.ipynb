```python
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
```

```python
orignal_img = cv.imread("UnLabelled/WhatsApp Image 2023-04-06 at 14.41.19.jpeg")
# img = cv.imread("trial_img/img.jpg")
img = orignal_img
display_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
print(f"Dimensions = {len(img)} x {len(img[0])}")
plt.imshow(display_img)
plt.show()
```

```python
rgb = cv.cvtColor(img, cv2.COLOR_BGR2RGB)
print("Current Image: ")
plt.imshow(rgb)
plt.show()

#TODO --> improve the lower and upper bound for color of the bacteria
# # lower bound and upper bound for Red color of bacteria it is in hsv
# # apparently

#first attempt
#horrible but got major bacteria hotspots
# lower_bound = np.array([0,0,0])
# upper_bound = np.array([300,245,300])

#second attempt
#seems to get confused between bacteria and blue voids
# lower_bound = np.array([0,0,0])
# upper_bound = np.array([300,155,300])

#third attempt
#switched to using rgb
#certainly better but loses out on single bacteria on the edges 
#further training could be used to enhance the accuracy
#But we can also adjust the lost in our average bacteria pixel density
#thereby enhancing the chances of a correct guess
lower_bound = np.array([70,0,0])
upper_bound = np.array([255,70,255])

# find the colors within the boundaries
mask = cv.inRange(rgb, lower_bound, upper_bound)

#Displaying the mask
print("Current mask")
plt.imshow(mask)
plt.show()
```

```python
#Removing unnecessary noise from masks:

#define kernel size  
kernel = np.ones((7,7),np.uint8)

# Remove unnecessary noise from mask

mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
```

```python
# Segment only the detected region
# Bitwise-AND mask and original image
print("Bitwise-AND mask and original image")
segmented_img = cv2.bitwise_and(img, img, mask=mask)
print("New image")
cv.imwrite("img_with_bacteria.jpg",segmented_img)
plt.imshow(cv.cvtColor(segmented_img,cv.COLOR_BGR2RGB))
plt.show()
```
